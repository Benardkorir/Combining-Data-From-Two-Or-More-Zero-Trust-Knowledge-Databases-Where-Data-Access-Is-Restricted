```python
#Define a funtion that runs on a specified driver and database,
#result as a graph or a DataFrame based on the 'return_graph' parameter
def run_query(driver, database_name, query, return_graph=False):
    with driver.session(database=database_name) as session:  
        result = session.run(query)
        # If 'return_graph' is True, return the result as a graph
        if return_graph:
            graph = result.graph()
            return graph
        # If 'return_graph' is False, return the result as a DataFrame
        else:
            # Convert the result to a DataFrame
            df = pd.DataFrame([r.values() for r in result], columns=result.keys())
            # If the DataFrame is empty, raise a ValueError
            if df.empty:
                raise ValueError("Access denied!")
            else:
                return df
```

```python
# Define styles for nodes to avoid defining in each line of code
node_styles = {
    "level1_data:product": {"color":"green","label":"productName"},
    "level2_data:product": {"color":"red","label":"productName"},
    "level1_data:category": {"color":"yellow","label":"categoryName"},
    "level2_data:category": {"color":"khaki","label":"categoryName"},
    "supplier:level1_data": {"color":"orange","label":"companyName"},
    "supplier:level2_data": {"color":"pink","label":"companyName"},
    "level2_data:order": {"color":"black","label":"orderID"},
    "level1_data:order": {"color":"bisque","label":"orderID"},
    "customer:level1_data": {"color":"venus","label":"customerID"},
    "level2_data:customer": {"color":"salmon","label":"customerID"},
    "level1_data:customer": {"color":"olive","label":"customerID"},
    "level2_data:shipper": {"color":"maroon","label":"companyName"},
    "level1_data:shipper": {"color":"darkviolet","label":"companyName"},
    "level1_data:employee": {"color":"green","label":"firstName"},
    "employee:level2_data": {"color":"indigo","label":"firstName"},
    "level1_data:territory": {"color":"pink","label":"territoryDescription"},
    "territory:level2_data": {"color":"grey","label":"territoryDescription"},
    "level1_data:region": {"color":"cyan","label":"regionDescription"},
    "level2_data:region": {"color":"olive","label":"regionDescription"},
    "DEFAULT": {"color":"purple","label":"label"}
}

# Define styles for relationships
relationship_styles = {
    "DEFAULT":{"color":"black","label":"label"}
}

# Function to create a graph for an airport
def createGraph(neo4jGraph):
    # Function to get the style of an item
    def get_style(styles_dict, item):
        return styles_dict.get(item["properties"]["label"], styles_dict["DEFAULT"])

    # Function to get the value of a style key for an item
    def get_style_value(styles_dict, item, key):
        style = get_style(styles_dict, item)
        return style.get(key, styles_dict["DEFAULT"][key])

    # Function to get the style of a relationship
    def get_relationship_style(relationship):
        return get_style(relationship_styles, relationship)

    # Function to get the label of an item based on its style
    def get_label(styles_dict, item):
        return item["properties"][get_style_value(styles_dict, item, "label")]

    w = GraphWidget(graph = graph)

    # Set mappings for nodes and edges in the graph widget
    w.set_edge_color_mapping(lambda index, edge : get_relationship_style(edge)["color"])
    w.set_node_styles_mapping(lambda index, node : node_styles.get(node["properties"]["label"], {}))
    w.set_node_label_mapping(lambda index, node : get_label(node_styles, node))
    w.set_edge_label_mapping(lambda index, edge : get_label(relationship_styles, edge))

    # Show the graph
    return w.show()
```

```python
# Now you can call this function with your query
graph = run_query(driver, database_name, 
                  """MATCH (p:product)-[r:PART_OF]->(c:category) 
                  RETURN p,c,r 
                  LIMIT 25""", return_graph=True)
# And visualize the result
createGraph(graph)
```
